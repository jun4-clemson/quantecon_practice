```python
ts_length = 100
ϵ_values = []   # empty list

for i in range(ts_length):
    e = np.random.randn()
    ϵ_values.append(e)

plt.plot(ϵ_values)
plt.show()
```

```python
ts_length = 100
ϵ_values = []
i = 0
while i < ts_length:
    e = np.random.randn()
    ϵ_values.append(e)
    i = i + 1
plt.plot(ϵ_values)
plt.show()
```

```python
# example of np.array.
# exercise calculate muture value
T= 50
r = 0.025
b=np.empty(T+1) 
b[0] = 10
for i in range(T): #(range(T) is 0 ,1, 2, ... , T-1)
    b[i+1]=b[i]*(1+r)

plt.plot(b,label="bank balance")
plt.legend()
plt.show()
# one advantage for np.array compare to list is we don't need to keep asking for new space
```

```python
T=200
α=0.9
x=np.empty(T+1)
for i in range(T):
    x[i+1]=x[i]*α+np.random.randn()
plt.plot(x)
plt.show()
```
